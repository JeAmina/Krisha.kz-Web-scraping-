```python
driver=webdriver.Chrome(executable_path=r'C:\Users\amina\anaconda3\Lib\site-packages\chromedriver_win32/chromedriver.exe')
```

```python
driver.maximize_window()
```

```python
driver.get('https://krisha.kz/prodazha/kvartiry/nur-sultan/')
```

```python
links=[]
number = True
t=1
n='Дальше'
while number:
    products = driver.find_elements_by_class_name('a-card__inc')
    for i in products:
        t=t+1
        i.find_elements_by_class_name('a-card__header-left')
        l=i.find_element_by_tag_name('a')
        links.append(l.get_property('href'))
        if t < 120:number=True
        else:number=False
    try:
        driver.find_element_by_link_text(n).click()
    except:
        number=False  
```

print(len(links))

```python
links
```

```python
all_info = []
for product in links:
    driver.get(product)

    city = driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div[1]/div[1]/div[2]/div[1]/div[3]/span').text.strip()
   
    name = driver.find_element_by_xpath('/html/body/main/div[2]/div/div[1]/h1').text.strip()
    
    floor = driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div[1]/div[1]/div[2]/div[3]/div[3]').text.strip()
    
    price = driver.find_element_by_class_name('offer__price').text.replace('&nbsp','').strip()
    
    take_phone = driver.find_element_by_xpath('//*[@id="tm-telephone-body"]').click()
    time.sleep(3)
    phone = driver.find_element_by_class_name('offer__contacts-phones').text.replace('\n',' ').strip()
        
    product = {
        'city':city,        
        'name':name,
        'floor':floor,
        'price':price,
        'phone':phone,
    }
    all_info.append(product)
```
